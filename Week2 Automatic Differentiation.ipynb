```python
import numpy as np
import matplotlib.pyplot as plt 

%matplotlib inline 

x = np.arange(0, 10, 0.1)

cos = np.cos(x)
min = -np.sin(x)

plt.plot(x, min)
plt.plot(x, cos)
```

```python
sin = np.sin(x)
max = np.cos(x)

plt.plot(x, max)
plt.plot(x, sin)
```

```python
import tensorflow as tf

x = tf.Variable(initial_value=5.0)
y = tf.cos(x)

trainstep = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(y)

with tf.Session() as sess: 
    writer = tf.summary.FileWriter('', sess.graph)
    writer.close()
```
