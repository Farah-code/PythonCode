```python
def DataAnalysis (features, labels): 
    from sklearn import linear_model 
    from sklearn.linear_model import LinearRegression
    from sklearn.linear_model import LogisticRegression 
    from sklearn  import preprocessing 
    import tabulate 
    from tabulate import tabulate
    import pandas_profiling as pp
    import matplotlib.pyplot as plt
    
    linear = LinearRegression().fit(features, labels)
    slop = linear.coef_
    intercept = linear.intercept_
    features_predict = linear.predict(features)
    
    accuracy = linear.score(features, features_predict)
    
    print(tabulate([['slop', slop], ['intercept', intercept],['predict Data', features_predict], ['accuracy', accuracy]], headers=['Linear Regression', 'Values']))
    

    
    
    mean = features_predict.mean()
    variance = features_predict.var()
    StandardDivation = features_predict.std()
    
    print(tabulate([['mean', mean], ['variance', variance], ['Standard Divation', StandardDivation]], headers=['quantative forcast ', 'Values']))
    

    encoder = preprocessing.LabelEncoder()
    labels = encoder.fit_transform(labels)
    
    logistic = LogisticRegression().fit(features,labels)
    
    features_predict = logistic.predict(features)
    accuracy = logistic.score(features, features_predict)
    
    
    print(tabulate([['predict Data', features_predict], ['accuracy', accuracy]], headers=['Logistic Regression', 'Values']))
    print('predict data' ,features_predict)
    print('accuarcy ' , accuracy)
    
    
    
    mean = features_predict.mean()
    variance = features_predict.var()
    StandardDivation = features_predict.std() 
    print(tabulate([['mean', mean], ['variance', variance], ['Standard Divation', StandardDivation]], headers=['quantative forcast ', 'Values']))
    
   
    plt.scatter(features_predict, labels)
    plt.title('Predict Data')
    plt.xlabel('labels')
    plt.ylabel('features_predict')
    plt.show()
```

```python
features = df[['Total Streams&Downloads (Academic Year)', 'Total Streams & Downloads (Prev. Year)']]
features
```

```python
labels = df['Marketing Tends Movment (Total Streams & Downloads)']
labels
```

```python
DataAnalysis (features, labels)
```
